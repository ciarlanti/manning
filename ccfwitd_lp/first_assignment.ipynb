```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
(trainset_reviews,trainset_polarity),(testset_reviews,testset_polarity) = tf.keras.datasets.imdb.load_data(
    path='imdb.npz',
    num_words=None,
    skip_top=0,
    maxlen=None,
    seed=113,
    start_char=1,
    oov_char=2,
    index_from=3,
)
```

```python
# without numpy.where()
positive_reviews_training = [r for i,r in enumerate(trainset_reviews) if trainset_polarity[i] == 1]
negative_reviews_training = [r for i,r in enumerate(trainset_reviews) if trainset_polarity[i] == 0]
positive_review_length = [len(each) for each in positive_reviews_training]
negative_review_length = [len(each) for each in negative_reviews_training]
```

```python
def output_for_assignment(data,lengths,kind): 
    max_length = max(set(lengths), key=lengths.count)
    print(f"The most frequent length for {kind} reviews is {max_length} words and it appears {lengths.count(max_length)} times")
```

```python
output_for_assignment(positive_reviews_training,positive_review_length,'positive')
output_for_assignment(negative_reviews_training,negative_review_length,'negative')
```

```python
sns.set_style('darkgrid')
fig,ax = plt.subplots()
sns.distplot(positive_review_length,label='Positive Reviews Lengths')
sns.distplot(negative_review_length,label='Negative Reviews Lengths')
ax.legend()
ax.set(xlabel='total word count', ylabel='frequency of occurrence')
```
